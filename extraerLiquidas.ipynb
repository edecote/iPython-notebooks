Create a DataAccess object to store all data found in path

```python
dataobj = da.DataAccess('Yahoo', s_datapath='/Users/enrique/qstk/mex/quotes', verbose=True)
```

```python
ls_all_syms = dataobj.get_all_symbols()
```

```python
d_start= dt.datetime(2012,01,01)
d_end=dt.datetime(2014,05,31)
timestamps = dumex.getBMVdays(d_start, d_end, dt.timedelta(hours=16))
ls_keys = ['open', 'high', 'low', 'close', 'volume', 'actual_close']
ldf_data = dataobj.get_data(timestamps, ls_all_syms, ls_keys)
d_data = dict(zip(ls_keys, ldf_data))
for s_key in ls_keys:
        d_data[s_key] = d_data[s_key].fillna(method='ffill')
        d_data[s_key] = d_data[s_key].fillna(method='bfill')
        d_data[s_key] = d_data[s_key].fillna(1.0)
```

```python
vol=d_data['volume']
```

```python
good_stocks = []
mean = vol.mean()
for sym in ls_all_syms:
    if mean[sym] >5000:
        good_stocks.append(sym)
        print 'sym:{0}, mean:{1}'.format(sym,mean[sym])
```

```python
file = open('liquidas.txt','w')
for s in good_stocks:
    file.write(s+'\n')
file.close()
```
