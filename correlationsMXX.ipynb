```python
import numpy as np
import QSTK.qstkutil.qsdateutil_mex as dumex
import QSTK.qstkutil.tsutil as tsu
import QSTK.qstkutil.DataAccess as da
import datetime as dt
import pandas as pd
```

```python
market = '^MXX'
dt_start = dt.datetime(2012,1,1)
dt_end = dt.datetime(2014,8,15)
dt_timeofday = dt.timedelta(hours=16)
c_dataobj = da.DataAccess('Yahoo', s_datapath='/Users/enrique/workspace/qstk/mex/quotes', verbose=True)
ls_symbols = c_dataobj.get_symbols_from_list('ETFs')
#ls_symbols.remove(market)
ldt_timestamps = dumex.getBMVdays(dt_start, dt_end, dt_timeofday)
```

```python
ls_keys = ['actual_close']
ldf_data = c_dataobj.get_data(ldt_timestamps, ls_symbols, ls_keys)
data = dict(zip(ls_keys, ldf_data))
d_data = data.copy()
for s_key in ls_keys:
        d_data[s_key] = d_data[s_key].fillna(method='ffill')
        d_data[s_key] = d_data[s_key].fillna(method='bfill')
        d_data[s_key] = d_data[s_key].fillna(1.0)
```

```python
data = d_data['actual_close']
na_data = data.values.copy()
tsu.returnize1(na_data)
naCov = np.cov(data,rowvar=False)
naCorr = np.corrcoef(data)
```
